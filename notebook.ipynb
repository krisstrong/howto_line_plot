# From Raw to Refined – A Step-by-Step Line Plot Transformation in Python
This notebook walks through the full process of improving a basic line plot using `matplotlib` — starting from a default plot, stripping it down to only the line, and incrementally refining it using better color, layout, and labels.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dummy monthly sales data
df = pd.DataFrame({
    'Month': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'Sales': [100, 120, 130, 90, 110, 160, 170, 150, 140, 180, 200, 210]
})

df.head()
```

## Step 1: Default Plot

This is the plot you get with minimal code. It’s functional but unpolished.

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Month'], df['Sales'])
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

## Step 2: Simplify the Plot

In this step, we will remove all non-essential elements, leaving only the data line. This minimalist approach ensures that the focus remains solely on the data, highlighting the upward trend even without additional context.

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(df['Month'], df['Sales'], color='black')

# Remove all visual clutter
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

ax.set_title('')
ax.set_xlabel('')
ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

## Step 3: Add Style and Meaning

We now start adding back structure with purpose:

- Custom color
- Axis ticks and labels
- A clear title and axis labels
- Thicker line

```python
fig, ax = plt.subplots(figsize=(8, 5))

# Styled line
ax.plot(df['Month'], df['Sales'], color='#1f77b4', linewidth=2.5)

# Show ticks
ax.set_xticks(df['Month'][::2])
ax.set_xticklabels(df['Month'][::2].dt.strftime('%b'), fontsize=10)
ax.set_yticks([100, 150, 200])
ax.set_yticklabels(['100', '150', '200'], fontsize=10)

# Add labels
ax.set_title('Monthly Sales in 2023', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Sales', fontsize=12)

# Style axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

The visualization has improved significantly, but there are still opportunities for further refinement.

## Step 4: Final Polish

We finalize with:

- Data point markers
- A subtitle or source note
- Better grid
- Fine-tuned layout

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data for demonstration
df = pd.DataFrame({
    'Month': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'Sales': [100, 120, 130, 90, 110, 160, 170, 150, 140, 180, 200, 210]
})

fig, ax = plt.subplots(figsize=(8, 5))

# Final line with different alpha for line and markers
ax.plot(df['Month'], df['Sales'], color='tab:blue', linewidth=2.5, alpha=0.7)
ax.plot(df['Month'], df['Sales'], linestyle='None', marker='o', markerfacecolor='tab:blue', markeredgewidth=1, markeredgecolor='white', markersize=10, alpha=1.0)

# Customize ticks
ax.set_xticks(df['Month'])
ax.set_xticklabels(df['Month'].dt.strftime('%b'), fontsize=10)
ax.tick_params(axis='x', which='both', length=0)
ax.tick_params(axis='y', which='both', length=0)

# Labels and title
ax.set_title('Monthly Sales – 2023', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('', fontsize=12)
ax.set_ylabel('Sales (USD)', fontsize=12)

# Grid & background
ax.grid(axis='y', linestyle='--', alpha=0.2)
ax.set_facecolor('#fafafa')

# Remove clutter
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Optional: Add a source note
plt.figtext(0.9, 0.01, 'Source: Dummy data', ha='right', fontsize=8, color='gray')

plt.tight_layout()
plt.show()
```

To enhance the visual focus of our line plot, we can adjust the line color and switch the plot's theme to a dark mode version.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data for demonstration
df = pd.DataFrame({
    'Month': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'Sales': [100, 120, 130, 90, 110, 160, 170, 150, 140, 180, 200, 210]
})

fig, ax = plt.subplots(figsize=(8, 5))

# Final line with different alpha for line and markers
ax.plot(df['Month'], df['Sales'], color='tab:orange', linewidth=3, alpha=0.7)
ax.plot(df['Month'], df['Sales'], linestyle='None', marker='o', markerfacecolor='tab:orange', markeredgewidth=1.5, markeredgecolor='#1E1E2E', markersize=10, alpha=1.0)

# Customize ticks
ax.set_xticks(df['Month'])
ax.set_xticklabels(df['Month'].dt.strftime('%b'), fontsize=10, color='white')
ax.tick_params(axis='x', which='both', length=0, colors='white')
ax.tick_params(axis='y', which='both', length=0, colors='white')

# Labels and title
ax.set_title('Monthly Sales – 2023', fontsize=16, fontweight='bold', pad=15, color='white')
ax.set_xlabel('', fontsize=12, color='white')
ax.set_ylabel('Sales (USD)', fontsize=12, color='white')

# Grid & background
ax.grid(axis='y', linestyle='--', alpha=0.5, color='#3A3F4A')
ax.set_facecolor('#1E1E2E')

# Remove clutter
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Optional: Add a source note
plt.figtext(0.99, 0.01, 'Source: Dummy data', ha='right', fontsize=8, color='gray')

# Set the figure background color
fig.patch.set_facecolor('#1E1E2E')
ax.set_facecolor('#1E1E2E')

plt.tight_layout()
plt.show()
```

Notice how the focus remains on the narrative and the direction it provides. The narrative now stands out clearly, ensuring that your Director will immediately grasp the plot and its significance.
